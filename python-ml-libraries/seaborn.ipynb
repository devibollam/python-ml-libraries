```python
import seaborn as sns
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd 
```

```python
# 1. Load the built-in 'tips' dataset
tips = sns.load_dataset("tips")
print("First 5 rows of tips dataset:\n", tips.head())

```

```python
# 2. Create a scatter plot of total_bill vs tip
plt.figure(figsize=(6,4))
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Total Bill vs Tip")
plt.show()
```

```python
# 3. Create a box plot for total_bill by day
plt.figure(figsize=(6, 4))
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Box Plot of total Bill by day")
plt.show()
```

```python
# 4. Create a histogram for the total_bill column
plt.figure(figsize=(6,4))
#sns.histplot(x="total_bill", data=tips)
sns.histplot(tips["total_bill"], bins=20, kde=True)
plt.title("Histogram of Total Bill")
plt.show()
```

```python
# 5. Create a count plot of the 'day' column
plt.figure(figsize=(6, 4))
sns.countplot(x="day", data=tips)
plt.title("Count of Days")
plt.show()
```

```python
# 6. Create a violin plot to show distribution of total_bill by day
plt.figure(figsize=(6, 4))
sns.violinplot(x="day", y="total_bill", data=tips)
plt.title("Violin Plot of Total Bill by Day")
plt.show()
```

```python
# 7. Create a heatmap of the correlation matrix
plt.figure(figsize=(6, 4))
corr_matrix = tips.select_dtypes(include=[np.number]).corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 8. Create a pair plot of the dataset
sns.pairplot(tips, hue="sex")
plt.show()
```
