```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1. Create a basic line plot
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.title("Basic Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
# 2. Scatter plot with random data
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
plt.scatter(x_scatter, y_scatter, color='red', marker='o')
plt.title("Scatter Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

```python
# 3. Bar chart with sample data
categories = ['A', 'B', 'C', 'D']
values_bar = [10, 20, 15, 25]
plt.bar(categories, values_bar, color='blue')
plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
# 4. Histogram with normal distribution data
data_hist = np.random.randn(1000)
plt.hist(data_hist, bins=30, color='green', alpha=0.7)
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# 5. Subplots with multiple charts
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs[0, 0].plot(x, y, color='blue')
axs[0, 0].set_title("Line Plot")

axs[0, 1].scatter(x_scatter, y_scatter, color='red')
axs[0, 1].set_title("Scatter Plot")

axs[1, 0].bar(categories, values_bar, color='purple')
axs[1, 0].set_title("Bar Chart")

axs[1, 1].hist(data_hist, bins=30, color='orange', alpha=0.7)
axs[1, 1].set_title("Histogram")

plt.tight_layout()
plt.show()
```

```python
# 6. Customize plots with grid, legends, and styles
plt.plot(x, y, label='Sine Wave', linestyle='--', linewidth=2, color='magenta')
plt.legend()
plt.grid(True)
plt.title("Customized Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
# 7. Save a plot as an image
plt.plot(x, y)
plt.title("Saving a Plot")
plt.savefig("plot.png")
plt.show()
```
